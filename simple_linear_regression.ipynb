## 講座21、22 模型理論

‧迴歸分析可做為評斷分類(變數間)的關聯程度<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;eg. 薪資和年資的相關性<br>
‧訪查、調研我們會知道年資越高薪資越高，但卻無法知道準確的量化程度<br>
‧利用簡單線性迴歸模型找到那條具最佳解釋力的線性組合<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;eg. 年資的增減如何影響薪資的結果<br>
‧簡單線性迴歸模型<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;y：相關變數、被解釋變數<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;eg. 薪資 vs. 工作年資、考試分數 vs. 用功程度<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;x1：獨立變數、解釋變數，內含假設x改變會影響y的變化<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;順帶一提，當然x變數有可能根本不是影響y的直接因子<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;(後面的課程會提到這個議題，變數的隱含關聯性)<br>

b1：x1的係數、斜率項，指x1變動一單位影響y的變化(正負、大小)<br>
b0：截距項，反應當x1=0時y的值<br>
eg. 新鮮人剛入社會的薪資行情<br>
